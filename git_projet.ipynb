```python
inventaires = {
    "bananes" : 120,
    "mangues" : 85,
    "ananas" : 45,
    "noix de coco" : 60,
    "papayes" : 30
}
```

```python
def affichier_inventaire(inventaires):
    print("inventaire actuel de la plantation :")
    for fruit, quantite in inventaires.items():
        print(f"- {fruit.capitalize()}: {quantite} unites")
```

```python
affichier_inventaire(inventaires)
```
